```python
#import Image  # PIL
import numpy
import pylab

from pywt import WaveletPacket2D

im = Image.open("letter.png").convert('L')
arr = numpy.fromstring(im.tobytes(), numpy.uint8)
arr.shape = (im.size[1], im.size[0])

wp2 = WaveletPacket2D(arr, 'db2', 'sym', maxlevel=2)

pylab.imshow(arr, interpolation="nearest", cmap=pylab.cm.gray)

path = ['d', 'v', 'h', 'a']

#mod = lambda x: x
#mod = lambda x: abs(x)
mod = lambda x: numpy.sqrt(abs(x))

pylab.figure()
for i, p2 in enumerate(path):
    pylab.subplot(2, 2, i + 1)
    p1p2 = p2
    pylab.imshow(mod(wp2[p1p2].data), origin='image', interpolation="nearest",
        cmap=pylab.cm.gray)
    pylab.title(p1p2)

for p1 in path:
    pylab.figure()
    for i, p2 in enumerate(path):
        pylab.subplot(2, 2, i + 1)
        p1p2 = p1 + p2
        pylab.imshow(mod(wp2[p1p2].data), origin='image',
            interpolation="nearest", cmap=pylab.cm.gray)
        pylab.title(p1p2)

pylab.figure()
i = 1
for row in wp2.get_level(2, 'freq'):
    for node in row:
        pylab.subplot(len(row), len(row), i)
        pylab.title("%s=(%s row, %s col)" % (
        (node.path,) + wp2.expand_2d_path(node.path)))
        pylab.imshow(mod(node.data), origin='image', interpolation="nearest",
            cmap=pylab.cm.gray)
        i += 1

pylab.show()
```

```python
import numpy
import pylab

from pywt import WaveletPacket

x = numpy.arange(612 - 80, 20, -0.5) / 150.
data = numpy.sin(20 * pylab.log(x)) * numpy.sign((pylab.log(x)))
from sample_data import ecg as data

wp = WaveletPacket(data, 'sym5', maxlevel=4)

pylab.bone()
pylab.subplot(wp.maxlevel + 1, 1, 1)
pylab.plot(data, 'k')
pylab.xlim(0, len(data) - 1)
pylab.title("Wavelet packet coefficients")

for i in range(1, wp.maxlevel + 1):
    ax = pylab.subplot(wp.maxlevel + 1, 1, i + 1)
    nodes = wp.get_level(i, "freq")
    nodes.reverse()
    labels = [n.path for n in nodes]
    values = -abs(numpy.array([n.data for n in nodes]))
    pylab.imshow(values, interpolation='nearest', aspect='auto')
    pylab.yticks(numpy.arange(len(labels) - 0.5, -0.5, -1), labels)
    pylab.setp(ax.get_xticklabels(), visible=False)

pylab.show()
```
